```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Setup
sns.set(style="whitegrid")
np.random.seed(101)
```

```python
# Generate exponential and uniform distributions
exp_data = np.random.exponential(scale=2.0, size=1000)
uniform_data = np.random.uniform(low=0, high=10, size=1000)

# Plot both using subplot
fig, axs = plt.subplots(1, 2, figsize=(14, 5))

sns.histplot(exp_data, kde=True, bins=30, color="salmon", ax=axs[0])
axs[0].set_title("Exponential Distribution")

sns.histplot(uniform_data, kde=True, bins=30, color="teal", ax=axs[1])
axs[1].set_title("Uniform Distribution")

plt.suptitle("Simulated Probability Distributions", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Generate Poisson data
poisson_data = np.random.poisson(lam=4, size=500)

# Boxplot
sns.boxplot(x=poisson_data, color='orchid')
plt.title("Boxplot of Poisson Distribution (=4)")
plt.xlabel("Value")
plt.show()
```

```python
# Assume sample from known population std
sample = np.random.normal(loc=70, scale=8, size=40)
pop_mean = 68
pop_std = 8

z_stat = (np.mean(sample) - pop_mean) / (pop_std / np.sqrt(len(sample)))
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print(f"Z-statistic: {z_stat:.4f}")
print(f"p-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject null hypothesis: sample mean significantly different from 68")
else:
    print("Fail to reject null hypothesis")
```

```python
# Simulate two independent groups (non-normal)
group_a = np.random.exponential(scale=2.0, size=100)
group_b = np.random.exponential(scale=2.5, size=100)

# Violin Plot
sns.violinplot(data=[group_a, group_b])
plt.xticks([0, 1], ['Group A', 'Group B'])
plt.title("Distribution of Two Exponential Samples")
plt.show()

# Mann-Whitney U Test
u_stat, p_val = stats.mannwhitneyu(group_a, group_b)

print(f"U-statistic: {u_stat}")
print(f"p-value: {p_val:.4f}")

if p_val < 0.05:
    print(" Distributions are significantly different")
else:
    print(" No significant difference between distributions")
```

```python
# Generate new data
normal_data = np.random.normal(loc=0, scale=1, size=1000)
uniform_data = np.random.uniform(low=-2, high=2, size=1000)

# Sort data for CDF
norm_sorted = np.sort(normal_data)
uniform_sorted = np.sort(uniform_data)

# Plot empirical CDF
plt.plot(norm_sorted, np.linspace(0, 1, len(norm_sorted)), label="Normal CDF", color='blue')
plt.plot(uniform_sorted, np.linspace(0, 1, len(uniform_sorted)), label="Uniform CDF", color='green')

plt.title("CDF Comparison: Normal vs Uniform")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.legend()
plt.grid(True)
plt.show()
```
